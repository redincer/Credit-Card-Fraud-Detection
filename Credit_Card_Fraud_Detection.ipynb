```python
x = credit.iloc[:, :-1] # Independe variables
y = credit.iloc[:, -1] # Dependent Varible
```

```python
from collections import Counter
from imblearn.combine import SMOTETomek
```

```python

os = SMOTETomek(n_jobs=-1)
X_resample, Y_resample = os.fit_sample(x, y)
print('Before: {}'.format(Counter(y)))
print('Before: {}'.format(Counter(Y_resample)))
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X_resample, Y_resample, test_size=0.2, random_state=100)
# checking the new shapes
print("Shape of x_train: ", x_train.shape)
print("Shape of x_test: ", x_test.shape)
print("Shape of y_train: ", y_train.shape)
print("Shape of y_test: ", y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
knn_clf_model = KNeighborsClassifier(weights='distance', n_jobs=-1)
knn_clf_model.fit(x_train, y_train)
```

```python
knn_y_pred = knn_clf_model.predict(x_test)

print('Accuracy Score: \n', accuracy_score(y_test, knn_y_pred))
print('Confusion Matrix: \n', confusion_matrix(y_test, knn_y_pred))
print('Classification Report: \n', classification_report(y_test, knn_y_pred))
print('AUC score is: ', roc_auc_score(y_test, knn_y_pred))
```
